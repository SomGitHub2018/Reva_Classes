```python
credit = pd.read_csv('credit.csv')
credit.head()
credit.info()
```

```python
#Removing unnecessary columns using iloc
credit = credit.iloc[:,1:]
credit.sample(10)
credit.info()
```

```python
#Create new variable in the df which is going to be independent variable
credit['Balance_Per_Card'] = credit['Balance']/credit['Cards']
credit.head(10)
```

```python
# Drop variables from df usinf drop method
credit = credit.drop(['Cards','Balance'],1)
credit.head(10)
```

```python
print('Column Names \n')
credit.columns
print('\n')
print('Description \n')
credit.describe()
print('\n')
print('Info \n')
credit.info()
print('\n')
print('Shape \n')
credit.shape
print('\n')
print('Non Missing Values \n')
credit.count()
print('\n')
print('Missing Values \n')
credit.isna().sum()
print('\n')
```

```python
for col in credit.columns:
    if credit[col].dtype == 'object':
        print(np.round(credit.groupby(col)['Balance_Per_Card'].mean(),2))
```

```python
for col in credit.columns:
    if credit[col].dtype == 'object':
        np.round(credit.groupby(col).mean(),2)
```

```python
sns.pairplot(data=credit, hue='Ethnicity', diag_kind="kde", palette ="Dark2")
plt.show()
```

```python
correlations = credit.corr()
mask = np.zeros_like(correlations)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(correlations, mask=mask, cmap="Blues", annot=True)
sns.set(rc={'figure.figsize':(15,15)})
plt.show()
```

```python
#Do dummy coding using for loop
for col in credit.columns:
    if credit[col].dtype=='object':
        credit[col]=pd.Categorical(credit[col]).codes
```

```python
credit.info()
```

```python
#Create features and labels for decision tree 
X = credit.drop(['Balance_Per_Card'],1)
Y = credit['Balance_Per_Card']
```

```python
X.head()
Y.head()
```

```python
from sklearn import tree
reg = tree.DecisionTreeRegressor(min_samples_leaf=30)
reg = reg.fit(X,Y)
```

```python
import graphviz
dot_data = tree.export_graphviz(reg, out_file=None,
                               feature_names=X.columns,
                               class_names=Y,
                               filled=True, rounded=True,
                               special_characters=True)
graph = graphviz.Source(dot_data)
graph
graph.render("credit")
```

```python
#Find out the predicted values using the tree
predict = reg.predict(X)
```
